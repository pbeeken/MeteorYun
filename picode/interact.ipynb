```python
aspectRatio = 11 / 36.  # multiplier to x ratio of y movement to x movement in stepper steps.
```

```python
# circle of radius r
#
def circle(radius):
    #lc.goto([radius,0])
    #time.sleep(0.3)
    lc.laserPower(0.1)
    for angle in np.linspace(0, 2*np.pi):
        x = int(2*radius * np.cos(angle) * aspectRatio )
        y = int(radius * np.sin(angle))
        rc = lc.goto([x, y])
        print(f"{rc}")
        time.sleep(0.1)

    #print(f"{lc.goto([0, 0])}")
    lc.laserPower(0.0)

for i in range(5):
    circle(400)
```

```python
def box(sze):
    hold = 1.0
    xsze = int(sze * aspectRatio)
    ysze = int(sze)
    lc.laserPower(0.1)
    for i in range(10):
        lc.goto([xsze, ysze])
        lc.getREADY()
        time.sleep(hold)
        lc.goto([xsze, -ysze])
        lc.getREADY()
        time.sleep(hold)
        lc.goto([-xsze, -ysze])
        lc.getREADY()
        time.sleep(hold)
        lc.goto([-xsze, ysze])
        lc.getREADY()
        time.sleep(hold)
        lc.goto([xsze, ysze])
        lc.getREADY()
        time.sleep(hold)
    lc.laserPower(0.0)

lc.reZeroAxes()
box(400)
```
